Real-World Problem: A company needs to estimate GPU memory usage.
Your Objective: Compare memory usage of image and video tensors.

What You Learn
•	Why video models are expensive
•	Resource planning


```python
import numpy as np
```

## Create Image Tensor

```python
img = np.random.rand(224, 224, 3)

print("Image shape:", img.shape)
print("Image bytes:", img.nbytes)
```

## create video Tensor

```python
vid = np.random.rand(30, 224, 224, 3)

print("Video shape:", vid.shape)
print("Video bytes:", vid.nbytes)
```

## Compare Image vs video

```python
print("Video uses", vid.nbytes // img.nbytes, "times more memory than a single image")
```

```python
# Deep Learning FrameWork Convert Tensor to float
# B/c DL model 32-bit folad which reduce Memory by Half
img32 = np.random.rand(224, 224, 3).astype(np.float32)
vid32 = np.random.rand(30, 224, 224, 3).astype(np.float32)

print("Image (float32) bytes:", img32.nbytes)
print("Video (float32) bytes:", vid32.nbytes)
```
