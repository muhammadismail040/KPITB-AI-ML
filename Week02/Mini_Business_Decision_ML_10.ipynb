Real-World Problem: An AI system flags fraud transactions.

```python
import numpy as np

# For reproducibility
np.random.seed(42)
```

Generate Fruad Probability

```python
# Simulate fraud probability scores for 50 transactions
scores = np.random.rand(50)

# Preview first 10 scores
scores[:10]
```

Define Thrashhold

```python
# Business decision thresholds
thresholds = [0.5, 0.7, 0.9]

#Interpretation:
#0.5 → aggressive (more fraud caught, more false alarms)
#0.9 → conservative (fewer alerts, higher fraud risk)
```

Compare alert for each Threshhold

```python
for t in thresholds:
    alerts = scores[scores > t]
    print(f"Threshold {t}: {len(alerts)} alerts")
```

## Busniss friendly Summary

```python
for t in thresholds:
    alerts = scores[scores > t]
    print(f"""
Threshold: {t}
Flagged Transactions: {len(alerts)}
Risk Level: {"High Detection / High Cost" if t==0.5 else "Balanced" if t==0.7 else "Low Cost / Higher Risk"}
""")
```

## Final Learning Reflection

```python
print("""
KEY TAKEAWAYS:
• Lower threshold → More fraud caught, higher operational cost
• Higher threshold → Fewer alerts, risk of missed fraud
• Threshold tuning is a BUSINESS decision, not just ML
""")
```
