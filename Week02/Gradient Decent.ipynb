```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Patiant Recory Score
data = [4.2, 5.1, 3.8, 7.2, 5.5, 6.0]
print("Original Data:", data)
```

# Automaticaly Summary Fuction

```python
def summarize_clinic(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return mean, std_dev

mean, std = summarize_clinic(data)
print("Mean:", mean)
print("Standard Deviation:", std)
```

# Logic Chek

```python
# Add Outlier
data.append(15.0)
print("Updated Data:", data)

mean_new, std_new = summarize_clinic(data)
print("New Mean:", mean_new)
print("New Standard Deviation:", std_new)
```

```python
dosage = np.array([10, 20, 30, 40, 50])
recovery = np.array([15, 25, 35, 45, 50])
```

```python
plt.scatter(dosage, recovery)
plt.xlabel("Dosage")
plt.ylabel("Recovery")
plt.title("Dosage vs Recovery")
plt.show()
```

# Prediction Line

```python
w = 0.1
predicted = w * dosage

plt.scatter(dosage, recovery)
plt.plot(dosage, predicted)
plt.xlabel("Dosage")
plt.ylabel("Recovery")
plt.title("Prediction Line (w = 0.1)")
plt.show()
```

```python
weights = [0.2, 0.5, 0.8, 1.0]

plt.scatter(dosage, recovery)

for w in weights:
    plt.plot(dosage, w * dosage, label=f"w = {w}")

plt.xlabel("Dosage")
plt.ylabel("Recovery")
plt.title("Multiple Prediction Lines")
plt.legend()
plt.show()
```
