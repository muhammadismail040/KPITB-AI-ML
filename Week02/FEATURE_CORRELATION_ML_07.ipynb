```python
import numpy as np
```

Create the Numerical Lists (Features)

```python
ads = [100, 200, 300, 400, 500]   # Advertising budget
sales = [10, 20, 30, 40, 50]     # Sales numbers
```

# Calculate Correlation

```python
correlation_matrix = np.corrcoef(ads, sales)
print("Correlation Matrix:\n", correlation_matrix)
```

Extract Correlation Value

```python
correlation_value = correlation_matrix[0, 1]
print("Correlation between Ads and Sales:", correlation_value)
```

## Interpretation

```python
if correlation_value > 0.7:
    print("Strong positive correlation: Increasing ads increases sales.")
elif correlation_value > 0:
    print("Weak positive correlation.")
elif correlation_value < 0:
    print("Negative correlation.")
else:
    print("No correlation.")
```
