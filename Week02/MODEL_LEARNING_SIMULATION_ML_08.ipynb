```python
# No heavy libraries needed, but this is standard practice
import math
```

Initaialize parameters

```python
# Initial weight (model parameter)
w = 20

# Learning rate (how fast the model learns)
lr = 0.1

print("Initial Weight:", w)
print("Learning Rate:", lr)
```

Gradient Decent Simulation

```python
# Simulating learning for 10 steps
for i in range(10):

    # Loss function (Mean Squared Error simplified)
    loss = w ** 2

    # Gradient of loss with respect to weight
    gradient = 2 * w

    # Update weight using gradient descent
    w = w - lr * gradient

    print(f"Step {i+1} | Weight: {w:.2f} | Loss: {loss:.2f}")
```

## üîç What Is Happening?

‚Ä¢ **Loss**: Measures how wrong the model is  
‚Ä¢ **Gradient**: Direction and speed to reduce loss  
‚Ä¢ **Learning Rate**: Controls step size  
‚Ä¢ **Weight Update**: Improves model gradually  

### Key Insight
The model **does not learn instantly** ‚Äî  
it improves **step by step**, which explains why training takes time in real ML systems.
