# import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data

```python
# Exercise time (hours)
X = np.array([1, 2, 3, 4, 5])

# Calories burned (target)
y = np.array([100, 200, 300, 400, 500])
```

# Define Loss Fuction

```python
def calculate_loss(weight):
    predictions = weight * X
    loss = np.mean((predictions - y) ** 2)
    return loss
```

# Define Gradient Decent

```python
def calculate_gradient(weight):
    predictions = weight * X
    gradient = 2 * np.mean((predictions - y) * X)
    return gradient
```

# Gradient Decent Function fixed Bug

```python
def gradient_descent(learning_rate, iterations=100):
    weight = 0.5
    loss_history = []

    for i in range(iterations):
        gradient = calculate_gradient(weight)

        # ✅ FIXED LINE (Bug was here)
        weight = weight - learning_rate * gradient

        loss = calculate_loss(weight)
        loss_history.append(loss)

    return loss_history
```

# Testing Learning Rates

Start With small Learning Rate (0.000001)

```python
loss_slow = gradient_descent(learning_rate=0.000001)
```

Large Learning Rate

```python
loss_fast = gradient_descent(learning_rate=0.1)
```

# Plot the Curve

```python
plt.figure()
plt.plot(loss_slow, label="Learning Rate = 0.000001")
plt.plot(loss_fast, label="Learning Rate = 0.1")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss Curve Comparison")
plt.legend()
plt.show()
```

Learning rate = 0.000001 → Too slow, loss decreases very gradually

Learning rate = 0.1 → Converges quickly, reaches minimum faster
