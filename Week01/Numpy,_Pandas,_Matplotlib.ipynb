# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Numpy Array

```python
# Create NumPy array from 1 to 20
arr = np.arange(1, 21)

# Compute statistics
mean_val = np.mean(arr)
max_val = np.max(arr)
std_val = np.std(arr)

print("Array:", arr)
print("Mean:", mean_val)
print("Maximum:", max_val)
print("Standard Deviation:", std_val)
```

# 5x5 random matrix

```python
# Create a 5x5 random matrix
matrix = np.random.rand(5, 5)

print("Original Matrix:\n", matrix)

# Replace values less than 0.5 with 0
matrix[matrix < 0.5] = 0

print("\nUpdated Matrix (values < 0.5 replaced with 0):\n", matrix)
```

# PAndas Tsk

```python
# Create sample data
data = {
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha", "Usman"],
    "Marks": [78, 85, 62, 90, 70],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("students.csv", index=False)

print("CSV file created successfully!")
```

# Load csv

```python
# Load CSV file
df = pd.read_csv("students.csv")

# Display first 5 rows
df.head()
```

```python
# Calculate average of numeric column (Marks)
average_marks = df["Marks"].mean()
print("Average Marks:", average_marks)
```

# Filter rows & save to new CSV

```python
# Filter students with Marks greater than 75
filtered_df = df.loc[df["Marks"] > 75]

# Save filtered data to new CSV
filtered_df.to_csv("high_marks_students.csv", index=False)

print("Filtered CSV saved successfully!")
filtered_df
```

# Matplotlib Task

# Line graph

```python
# Values from 1 to 10
x = np.arange(1, 11)
y = x  # simple line

# Plot
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Graph from 1 to 10")
plt.show()
```

# Bar chart

```python
# Categories and values
categories = ["A", "B", "C", "D", "E"]
values = [10, 25, 15, 30, 20]

# Bar chart
plt.bar(categories, values)
plt.xticks(categories)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```
