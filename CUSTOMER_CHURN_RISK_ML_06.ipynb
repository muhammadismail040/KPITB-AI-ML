```python
# TASK 3: CUSTOMER CHURN RISK (PROBABILITY)
# Real-World Problem:
# A telecom company wants to predict which customers are likely to churn
# using probability outputs from a model.
```

```python
import numpy as np
import pandas as pd
```

# Simulate mkodel prediction

```python
# Simulating churn probability predictions for 20 customers
# Each value is between 0 and 1
np.random.seed(42)  # for reproducibility

predictions = np.random.rand(20)
predictions
```

DataFrame

```python
# Create customer IDs
customer_ids = [f"CUST_{i+1}" for i in range(20)]

# Create DataFrame
df = pd.DataFrame({
    "Customer_ID": customer_ids,
    "Churn_Probability": predictions
})

df
```

Define churn risk threshold

```python
# Threshold for high churn risk
threshold = 0.7
```

Identify High Risk Costumers

```python
# Filter high-risk customers
high_risk_customers = df[df["Churn_Probability"] > threshold]

high_risk_customers
```

```python
# Count High Risk cumstomers

print("Total customers:", len(df))
print("High-risk customers:", len(high_risk_customers))
```

Add Risk Label

```python
# Label customers based on risk
df["Churn_Risk"] = np.where(
    df["Churn_Probability"] > threshold,
    "High Risk",
    "Low Risk"
)

df
```

# Busnsiss insight Summary

```python
high_risk_percentage = (len(high_risk_customers) / len(df)) * 100

print(f"{high_risk_percentage:.2f}% of customers are at high risk of churn.")
```
